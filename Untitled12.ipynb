Aqui carregamos a foto que está em meu drive, e passamos para a rede, no qual vai observar e fazer a detecção e classificar.

```python
from ultralytics import YOLO

# Carrega o modelo (o 'n' é de nano, mais leve, e mais rápido)
modelo = YOLO('yolov8n.pt')

# Executa a detecção na sua imagem
# save=True cria uma pasta 'runs/detect/predict' com a foto marcada
results = modelo.predict(source='/content/drive/MyDrive/mulher com cachorro.avif', show=True, save=True)

# Se quiser ver o que foi detectado no console:
for r in results:
    print(r.boxes.cls) # Exibe os IDs das classes encontradas
```

Vamos analisar a saída do modelo YOLO:


O modelo processou a imagem, logo após a imagem foi redimensionada para 640x448 pixels para o processamento.
Foram detectados 1 pessoa e 3 cães (dogs) na imagem.
O tempo de inferência para essa imagem foi de 416.1ms (milissegundos).
Speed: 15.4ms preprocess, 416.1ms inference, 51.7ms postprocess ...: Isso detalha o tempo gasto em cada etapa do processo:

preprocess: tempo para preparar a imagem.


inference: tempo que o modelo levou para fazer a detecção em si.


postprocess: tempo para processar os resultados (como desenhar as caixas delimitadoras).


Results saved to: Isso significa que uma nova imagem, com as caixas delimitadoras e os rótulos dos objetos detectados, foi salva no caminho..... Você pode navegar até essa pasta no explorador de arquivos do Colab para ver a imagem resultante.

Tensor([16., 0., 16., 16.]): Este é o resultado do print(r.boxes.cls). Ele mostra as IDs das classes detectadas. No modelo YOLOv8n, que é treinado no conjunto de dados COCO:


0 corresponde a 'person' (pessoa).

16 corresponde a 'dog' (cão).

Portanto, o tensor indica que o modelo detectou uma pessoa (ID 0) e três cães (ID 16, 16, 16) na imagem, o que confirma o que foi impresso na linha de detecção.

Para deixar o trabalho mais completo decidi mostrar como a rede classificou e detectou a imagem apresentada, utilizando o display do python.

```python
import os
from IPython.display import Image, display

results_dir = '/content/runs/detect/predict'

# Listar arquivos na pasta de resultados
image_files = [f for f in os.listdir(results_dir) if f.endswith(('.jpg', '.jpeg', '.png', '.avif'))]

if image_files:
    # Assumindo que a imagem mais recente é a que queremos exibir
    # Ou, se houver apenas uma, será essa
    image_path = os.path.join(results_dir, image_files[0])
    print(f"Exibindo a imagem: {image_path}")
    display(Image(filename=image_path))
else:
    print(f"Nenhuma imagem encontrada em {results_dir}")
```

import os: Importa o módulo os, que fornece funções para interagir com o sistema operacional, como listar diretórios e manipular caminhos de arquivos.

from IPython.display import Image, display: Importa as classes Image e display do módulo IPython.display.

A função display é usada para renderizar objetos, como imagens, diretamente na saída do notebook Colab, e Image é a classe que representa a imagem a ser exibida.
