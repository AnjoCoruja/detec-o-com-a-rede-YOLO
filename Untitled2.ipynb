BIBLIOTECAS


```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
```

Vamos carregar a imagem de varias faces para detecção em uma rede já treinada, para que ela identifique os rostos e os marque, por meio do fatiamento de imagem.

```python

# 1. Carrega o modelo (o 'yolov8n-face' é otimizado para rostos)
# Se não tiver o peso de face, o 'yolov8n.pt' comum quebra o galho para detectar 'person'
model = YOLO('yolov8n')

# 2. Carrega a sua imagem
img_path = 'celebridades.png'
img = cv2.imread(img_path)

# 3. Roda a detecção
results = model(img)

# 4. Itera sobre as faces detectadas
for r in results:
    for box in r.boxes:
        # Coordenadas
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        conf = box.conf[0]

        # Desenha o retângulo em cada celebridade detectada
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img, f'{conf:.2f}', (x1, y1-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# 5. Mostra o resultado final com todos os quadrados
cv2_imshow(img)
```
